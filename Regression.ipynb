```python
import numpy as np
x ,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,
                           n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

```python

def step(z):
    return 1 if z>0 else 0
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

```python
intercept_,coef_ = perceptron(x,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
import numpy as np

# Step activation function
def step(x):
    return 1 if x >= 0 else 0

# Perceptron function
def perceptron(x, y, lr=0.01, epochs=10):
    # Add bias term to input data
    x = np.c_[np.ones(x.shape[0]), x]  # Add a column of ones for the bias
    weights = np.zeros(x.shape[1])  # Initialize weights
    m, b = [], []  # To store slopes and intercepts (optional)

    for epoch in range(epochs):
        for j in range(x.shape[0]):
            y_hat = step(np.dot(x[j], weights))
            weights = weights + lr * (y[j] - y_hat) * x[j]

        # Slope (m) and Intercept (b) for decision boundary visualization
        m.append(-(weights[1] / weights[2]))  # weights[1] = slope, weights[2] = weight for second feature
        b.append(-(weights[0] / weights[2]))  # weights[0] = bias term

    return m, b, weights

# Example usage
x = np.array([[2, 3], [1, 4], [3, 5]])  # Input data (3 samples, 2 features)
y = np.array([1, 0, 1])  # Labels

m, b, final_weights = perceptron(x, y)
print("Final Weights:", final_weights)
```
